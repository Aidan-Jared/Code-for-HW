```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def S(x, n):
    G = np.array([])
    for X in x:
        s = 0
        for m in range(n):
            a = 2 * ((1-np.cos((m+1) * np.pi))/((m+1) * np.pi)**2) * np.cos((m+1) * np.pi * X)
            s += a
        G = np.append(G,s)
    return G
x=np.linspace(-1,1,1000)
def f(x, n):
    return .5 + S(x,n)
```

```python
plt.plot(x, f(x, 1))
plt.plot(x, 1+x)
plt.plot(x, 1-x)
plt.ylim(0,1)
```
