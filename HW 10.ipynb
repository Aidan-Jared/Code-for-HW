```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w = lambda m: (2.*m + 1) * np.pi * 1. / 8.
k = lambda m: (2.*m + 1) * np.pi / 8.
B = lambda m: (-1.)**m * 32. / ((2.*m+1)**2 * np.pi**2) * (1. - np.cos((2.*m+1)*np.pi/8.))
```

```python
def F(x,t,n):
    g = np.array([])
    for X in x:
        s = 0
        for m in range(n):
            a = B(m)*np.sin(k(m)*X)*np.cos(w(m)*t)
            s += a
        g = np.append(g,s)
    return g
```

```python
x = np.linspace(0,8,1000)
plt.plot(x,F(x,0,5))
plt.title('t = 0', fontsize = 18)
```

```python
T = 2 * np.pi / w(0)
plt.plot(x,F(x,T/8,5))
plt.title('t = T/8', fontsize = 18)
```

```python
k = lambda n: (2.*n + 1)*np.pi/(2*1)
x = np.linspace(0,1,100)
def F(x,n):
    return np.sin(k(n)*x)
plt.plot(x,F(x,1),'r--',label = 'first mode')
plt.plot(x,F(x,2),'r-',label = 'second mode')
plt.plot(x,F(x,3),'r.',label = 'third mode')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., prop={'size':20})
```
