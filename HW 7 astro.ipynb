```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
c=3e5
H=71.
z=np.arange(0,.51,.01)
def dh1(z):
    return (c * z) / H
def dh2(z):
    return c/H * ((z+1)**2 - 1)/((z+1)**2 + 1)
def dl(z):
    return c/H * (1+z) * (z + 0.2025 * z**2 - 0.0529877 * z**3 + 0.04775 * z**4)
def dp(z):
    return c/H * (z + 0.2025 * z**2 - 0.0529877 * z**3 + 0.04775 * z**4)
plt.plot(z,dh1(z), ':',label='hubble distance')
plt.plot(z,dh2(z), '-',label='hubble distance, relatavistic')
plt.plot(z,dl(z), 'o',label='luminosity distance')
plt.plot(z,dp(z), '--',label='proper distance')
plt.xlabel('z')
plt.ylabel('c/H')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```
