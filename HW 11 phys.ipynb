```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def A(x):
    return 1 / ((1-x) + 4 * .05**2 * x**2)
w=np.arange(0,25,.5)
plt.plot(w,A(w),'-')
plt.xlabel('frequency')
plt.ylabel('amplitude (m)')
```

```python
def d(x):
    return np.arctan((.1 * x)/(1 - x**2))
w=np.arange(0,1,.01)
plt.plot(w,d(w),'-')
plt.xlabel('frequency')
plt.ylabel('delta')
```

```python
def X(t):
    return (-1) * np.e**(-.05 * t) * 10 * np.sin(t) + 10 * np.cos(t - np.pi / 2)
t = np.arange(0,20,.1)
plt.plot(t, X(t),'-')
plt.xlabel('time (s)')
plt.ylabel('position (m)')
```
